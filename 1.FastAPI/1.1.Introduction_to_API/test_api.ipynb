```python
import json
import requests
```

### API 1: Numbers

```python
# Status Code 404 - Not Found
response = requests.get("http://numbersapi.com/abcd")
response.status_code
```

```python
# Status Code 400 - Bad Request
response = requests.get("http://numbersapi.com/42.1")
response.status_code
```

```python
# Status Code 200 - OK
response = requests.get("http://numbersapi.com/42")
response.status_code
```

```python
response.text
```

### API 2: Weather

```python
# Status Code 503 - Service Unavailable
# The server cannot handle the request (because it is overloaded or down for maintenance). 
# Generally, this is a temporary state.
response = requests.get("https://goweather.herokuapp.com/weather/Mashhad")
response.status_code
```

```python
# Status Code 200 - OK
response = requests.get("https://goweather.herokuapp.com/weather/Mashhad")
response.status_code
```

```python
response.text
```

```python
json.loads(response.text)
```

### API 3: Lorem Picsum

```python
# Status Code 404 - Not Found
response = requests.get("https://picsum.photos/200/300/200")
response.status_code
```

```python
# Status Code 400 - Bad Request
response = requests.get("https://picsum.photos/10000000/1000000")
response.status_code
```

```python
# Status Code 200 - OK
response = requests.get("https://picsum.photos/200/300")
response.status_code
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt


file_data = np.frombuffer(response.content, dtype="int8")
image = cv2.imdecode(file_data, flags=cv2.IMREAD_UNCHANGED)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```
